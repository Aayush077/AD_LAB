#1.Read the student marks from the student_marks_10.csv file and perform the following: a) Calculate Total Marks and Average Marks for each student b) Add a column Result (Pass if Average ≥ 50, otherwise Fail)


```python
import pandas as pd
df = pd.read_csv("student_marks_10.csv")
df["Total Marks"] = df["Maths"] + df["Science"] + df["English"]

df["Average Marks"] = df["Total Marks"] / 3
df["Result"] = "Fail"
df.loc[df["Average Marks"] >= 50, "Result"] = "Pass"

print(df)
```

#2.Read employee details from the employee_salary_10.csv file and: a) Increase salary by 10% if experience ≥ 5 years b) Increase salary by 5% otherwise c) Add a column New_Salary

```python
import pandas as pd

df = pd.read_csv("employee_salary_10.csv")
df["New_Salary"] = df["Salary"]

df.loc[df["Experience"] >= 5, "New_Salary"] = df["Salary"] * 1.10
df.loc[df["Experience"] < 5, "New_Salary"] = df["Salary"] * 1.05

print(df)
```

#3.Read sales data from the sales_data_10.csv file and: a) Calculate total sales per product b) Calculate average sales per product


```python
import pandas as pd

df = pd.read_csv("sales_data_10.csv")
result = df.groupby("Product")["Sales"].agg(
    Total_Sales="sum",
    Average_Sales="mean"
)

print(result)
```

#4. Read student marks from the missing_marks_10.csv file and: a) Identify missing values b) Replace missing values with the mean of the Marks column Sample Input: missing_marks_10.csv

```python
import pandas as pd

df = pd.read_csv("missing_marks_10.csv")

print("Missing values:\n", df.isnull())

mean_marks = df["Marks"].mean()

df["Marks"] = df["Marks"].fillna(mean_marks)

print("\nAfter replacing missing values:\n", df)
```
