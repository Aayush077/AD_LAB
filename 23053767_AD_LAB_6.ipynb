#1. Write a Pandas program to add, subtract, multiple and divide two Pandas Series.:

```python
import pandas as pd

s1 = pd.Series([10, 20, 30, 40])
s2 = pd.Series([2, 4, 6, 8])

add = s1 + s2
print("Addition:")
print(add)

sub = s1 - s2
print("\nSubtraction:")
print(sub)

mul = s1 * s2
print("\nMultiplication:")
print(mul)

div = s1 / s2
print("\nDivision:")
print(div)
```

#4. Write a Pandas program to convert a given Series to an array.

#5. Write a Pandas program to convert Series of lists to one Series.

```python
import pandas as pd

series = pd.Series([[1, 2, 3], [4, 5], [6, 7, 8]])

new_series = series.explode()

print("Single Pandas Series:")
print(new_series)
```

#9. Write a Pandas program to select the rows where the score is missing, i.e. is NaN. Sample Python dictionary data and list labels:

#10. Pandas program to append a new row 'k' and then delete it

```python
import pandas as pd
import numpy as np

exam_data = {
    'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily',
             'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
    'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
    'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
    'qualify': ['yes', 'no', 'yes', 'no', 'no',
                'yes', 'yes', 'no', 'no', 'yes']
}

labels = ['a','b','c','d','e','f','g','h','i','j']

df = pd.DataFrame(exam_data, index=labels)

df.loc['k'] = ['Suresh', 15.5, 2, 'yes']
print("After adding new row:")
print(df)

df = df.drop('k')
print("\nAfter deleting new row:")
print(df)
```
