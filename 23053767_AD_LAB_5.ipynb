#8. Write a program to create a 4x4 matrix with values 1,2,3,4 just below & above the diagonal, rest zeros.

```python
import numpy as np

arr1 = np.zeros((4, 4), dtype=int)

values = [1, 2, 3, 4]

for i in range(4):
    if i < 3:
        arr1[i, i+1] = values[i]

for i in range(4):
    if i > 0:
        arr1[i, i-1] = values[i]

print(arr1)
```

#.9. Write a program to extract all the contiguous 3x3 blocks from a random 10x10 matrix.

```python
import numpy as np

arr1 = np.random.randint(1, 5, (10, 10))

blocks = []

for row in range(8):  # 10 - 3 + 1 = 8
    for col in range(8):
        arr = arr1[row:row+3, col:col+3]
        blocks.append(arr)

print("3x3 blocks:")
for index, arr in enumerate(blocks):
    print(f"Block {index + 1}:")
    print(arr)
    print()

print("Original array:")
print(arr1)
```

# 10. A magic square is a matrix all of whose row sums, column sums and the sums of the two diagonals are the same. (One diagonal of a matrix goes from the top left to the bottom right, the other diagonalgoesfrom top right to bottom left.) Show by direct computation that if the matrix A


```python
import numpy as np

A = np.array([
    [17, 24,  1,  8, 15],
    [23,  5,  7, 14, 16],
    [ 4,  6, 13, 20, 22],
    [10, 12, 19, 21,  3],
    [11, 18, 25,  2,  9]
])

row_sums = np.sum(A, axis=1)

col_sums = np.sum(A, axis=0)

main_diag = np.sum(np.diag(A))
other_diag = np.sum(np.diag(np.fliplr(A)))

all_sums = np.concatenate((row_sums, col_sums, [main_diag, other_diag]))

if np.max(all_sums) == np.min(all_sums):
    print("A is a magic square.")
    print("Magic sum =", np.max(all_sums))
else:
    print("A is NOT a magic square.")
```
